Copy prepared triplet data from Google Drive into the repository data folder.

```python
import shutil
from pathlib import Path

drive_triplets = Path('/content/drive/MyDrive/data/triplets')  # update if needed
local_triplets = Path('/content/vfi-app/data/triplets')
local_triplets.parent.mkdir(parents=True, exist_ok=True)
if local_triplets.exists():
    shutil.rmtree(local_triplets)
shutil.copytree(drive_triplets, local_triplets)
print(f'Copied triplets to: {local_triplets}')
```

Verify that a GPU is available before starting training.

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'No GPU found')
```

Configure and run model training.

```python
from src.train import train

config = {
    'data_dir': 'data/triplets',
    'epochs': 50,
    'batch_size': 8,
    'learning_rate': 1e-4,
    'checkpoint_dir': 'models',
    'device': 'cuda'
}

train(config)
```

Copy the best model checkpoint from Colab back to Google Drive for long-term storage.

```python
from pathlib import Path
import shutil

best_model = Path('/content/vfi-app/models/best.pth')
drive_model = Path('/content/drive/MyDrive/vfi-models/best.pth')
drive_model.parent.mkdir(parents=True, exist_ok=True)
shutil.copy2(best_model, drive_model)
print(f'Best model copied to: {drive_model}')
```
